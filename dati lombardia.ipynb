```python
import pandas as pd
```

```python
elenco_dataset='https://www.dati.lombardia.it/api/views/425r-pyq4/rows.csv?accessType=DOWNLOAD'
```

```python
datasets = pd.read_csv(elenco_dataset)
```

```python
datasets.columns
```

```python
datasets.TIPO.unique()
```

```python
datasets.groupby(['TIPO']).size().to_frame('size').reset_index()
```

```python
tablesids_dataset = datasets[datasets.TIPO=='dataset'].TABLE_ID.unique()
```

```python
tablesids = datasets.TABLE_ID.unique()
```

```python
len(tablesids)-len(tablesids_dataset)
```

```python
type(tablesids)
```

```python
diffs = list(set(tablesids) - set(tablesids_dataset))
```

```python
tableid_withoutdataset = datasets[datasets.TABLE_ID.isin(diffs)]
```

```python
tableid_withdataset =  datasets[~datasets.TABLE_ID.isin(diffs)]
```
